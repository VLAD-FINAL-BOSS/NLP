```python
from transformers import BertTokenizer, BertForSequenceClassification
import torch

from google.colab import drive
drive.mount('/content/drive')
# Путь к сохраненной модели
model_path = "/content/drive/MyDrive/bert_toxic_classifier"

# Загружаем модель и токенизатор
model = BertForSequenceClassification.from_pretrained(model_path)
tokenizer = BertTokenizer.from_pretrained(model_path)

# Перевод модели на устройство
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
model.eval()
print("Модель загружена и готова к проверке.")
```

```python
def classify_comment(text):
    # Токенизация текста
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=128)

    # Перенос тензоров на устройство (GPU/CPU)
    inputs = {key: val.to(device) for key, val in inputs.items()}

    # Получение предсказаний
    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits
        predicted_class = torch.argmax(logits, dim=1).item()

    return "Плохой комментарий" if predicted_class == 1 else "Хороший комментарий"


```

```python
# Примеры теста
print(classify_comment("Ты тупой"))
print(classify_comment("Спасибо за вашу помощь."))
```
